### Created a parser for the model as its need data to be sentence form

```python
text = text.lower() # converted to lower case so that capital are considered to be same as small letter
parser = PlaintextParser.from_string(text, Tokenizer('english'))
```

```python
summarizer_lex = LexRankSummarizer() # creating object of class
```

Cool right, so summary now is given based on sentences and they are directly been taken here let give same input to different models also and see what they give.

LSA (Latent semantic analysis) : It combines frequency technique with singular value decomposition to summarize texts

```python
from sumy.summarizers.lsa import LsaSummarizer
las_summarizer = LsaSummarizer()
summary = las_summarizer(parser.document, 2)
for sentence in summary:
  print(sentence)
```

Now summary is different and somewhat more correct then LexRank

Now try **textRank**

```python
from sumy.summarizers.text_rank import TextRankSummarizer
textrank_summarizer = TextRankSummarizer()
summary = textrank_summarizer(parser.document, 2)
for sentence in summary:
  print(sentence)
```
